```python
import re
def afficher_contextes(chaine , terme , taille_contexte = 30):
    match = re.search(terme , chaine)
    contexts = []
    while match is not None:
        #Les b o r n e s g a u c h e e t d r o i t e a u t o u r du mot :
        gauche = max(match.start() - taille_contexte - 1, 0)
        droite = match.end()+1+ taille_contexte
        contexts . append(chaine[gauche:droite])
        chaine = chaine[match.end():]
        match = re. search (terme , chaine )
    for c in contexts :
        print (c)
```

```python
# Calcul de TF
def creer_TF(chemin):
    allDocs = {}
# chemin = 'europarl/fr'
    chemin_fichier = glob.glob(chemin + '/*.txt') # [:5]
    for f in chemin_fichier:
        doc_val = {}
        nb_mots_doc = 0
        with open(f,'r') as fichier:  
            for ligne in fichier:        
                for mots in ligne.split():
                    nb_mots_doc += 1
                    if mots not in doc_val:
                        doc_val[mots] = 1
                    else:
                        doc_val[mots] += 1
        # print(nb_mots_doc)
        for k in doc_val:
            doc_val[k] /= nb_mots_doc
        allDocs[f] = doc_val
    return allDocs
# print(allDocs)                
    
```

```python
tf = creer_TF('europarl/fr')
print('les clés des vecteurs de mots par docs sont :\n', list(tf.keys())[:5])
print('les valeurs des vecteurs de mots par docs sont :\n', list(tf.values())[:5])
```

```python
import math as m
# Calcul de IDF
def creer_idf(index, index_inverse):
    mot2idf = {}
    for cle in index:
        mot2idf[cle] = m.log(len(index_inverse.keys())/len(index[cle]))
    return mot2idf

# print(mot2idf)
```

```python
idf = creer_idf(index, index_inverse)
print('les clés des vecteurs de mots par docs sont :\n', list(idf.keys())[:5])
print('les valeurs des vecteurs de mots par docs sont : \n', list(idf.values())[:5])
```

```python
def creer_tfidf(tf, idf):
    tfidf = {}
    dic = {}
    for cle1 in tf:
        for cle2 in mot2idf:
            if cle2 in tf[cle1]:
                dic[cle2] = mot2idf[cle2] * tf[cle1][cle2]
            else:
                dic[cle2] = 0
        tfidf[cle1] = dic
    return tfidf
```

```python
tfidf = creer_tfidf(tf, idf)
print('les clés des vecteurs de mots par docs sont :\n', list(tfidf.keys())[:2])
print('les valeurs des vecteurs de mots par docs sont : \n', list(tfidf.values())[:2])
```

```python
def tf_requete(requete):
    allDocs = {}
    doc_val = {}
    for mots in requete.split():
        if mots not in doc_val:
            doc_val[mots] = 1
        else:
            doc_val[mots] += 1
        # print(nb_mots_doc)
        for k in doc_val:
            doc_val[k] /= len(requete.split())
    allDocs['doc'] = doc_val
    return allDocs
```

```python
def calculer_sim_cosinus(docA, docB):
    norm_vec = 0
    scalaire_2_vec = 0
    norm_vec_doc = 0
    for cle1 in docA:
        norm_vec_doc += (docA[cle1] ** 2)
        norm_vec += (docB[cle1] ** 2)
        scalaire_2_vec += docA[cle1] * docB[cle1]
    sim_cos = scalaire_2_vec / (m.sqrt(norm_vec_doc) * m.sqrt(norm_vec))
    return sim_cos

def vect_sim_cosinus_doc(tfidf, tfidf_req):
    dic = {}
    for cle in tfidf:
        sim_cos = calculer_sim_cosinus(tfidf[cle], tfidf_req['doc'])
        dic[cle] = sim_cos
    return dict(sorted(dic.items(), key=lambda item: item[1]))

documents_retrouves = vect_sim_cosinus_doc(tfidf, tfidf_req)
        
        
```
